```python
import gym
from gym import spaces
import numpy as np
import random
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
from stable_baselines3 import PPO  # You can use any RL agent from stable-baselines3


# Task generation function
def generate_task(task_id):
    task_types = {
        "image_classification": {
            "gpu_demand": (20, 40), "memory_demand": (10, 30), "duration": (5, 10), "priority": (1, 3)
        },
        "lstm_training": {
            "gpu_demand": (30, 50), "memory_demand": (40, 60), "duration": (7, 15), "priority": (2, 4)
        },
        "cnn_training": {
            "gpu_demand": (50, 80), "memory_demand": (30, 50), "duration": (10, 20), "priority": (3, 5)
        }
    }
    task_type = random.choice(list(task_types.keys()))
    params = task_types[task_type]
    gpu_demand = random.randint(*params["gpu_demand"])
    memory_demand = random.randint(*params["memory_demand"])
    duration = random.randint(*params["duration"])
    priority = random.randint(*params["priority"])

    return {
        "id": task_id,
        "type": task_type,
        "gpu_demand": gpu_demand,
        "memory_demand": memory_demand,
        "duration": duration,
        "priority": priority,
        "remaining_duration": duration,
        "start_time": None
    }


# Environment class
class MLWorkloadSchedulingEnv(gym.Env):
    def __init__(self, num_gpus=4):
        super(MLWorkloadSchedulingEnv, self).__init__()
        self.num_gpus = num_gpus
        self.gpu_capacity = [100] * num_gpus
        self.memory_capacity = [100] * num_gpus

        self.observation_space = spaces.Box(low=0, high=100, shape=(2 * num_gpus,), dtype=np.float32)
        self.action_space = spaces.Discrete(num_gpus)

        self.tasks = []
        self.gpu_usage = [0] * num_gpus
        self.memory_usage = [0] * num_gpus
        self.total_time = 0
        self.task_waiting_times = []
        self.running_tasks = []
        self.completed_tasks = []
        self.gpu_usage_over_time = [[] for _ in range(num_gpus)]  # Track GPU usage over time

    def reset(self):
        num_tasks = random.randint(40,50)
        self.tasks = [generate_task(i) for i in range(num_tasks)]
        self.tasks.sort(key=lambda x: -x["priority"])
        self.gpu_usage = [0] * self.num_gpus
        self.memory_usage = [0] * self.num_gpus
        self.total_time = 0
        self.task_waiting_times = [0] * len(self.tasks)
        self.running_tasks = []
        self.completed_tasks = []
        self.gpu_usage_over_time = [[] for _ in range(self.num_gpus)]
        return np.array(self.gpu_usage + self.memory_usage, dtype=np.float32)

    def step(self, action):
        reward = 0

        # Process running tasks
        for task in self.running_tasks[:]:
            task["remaining_duration"] -= 1
            if task["remaining_duration"] <= 0:
                self.gpu_usage[task["gpu_id"]] -= task["gpu_demand"]
                self.memory_usage[task["gpu_id"]] -= task["memory_demand"]
                self.running_tasks.remove(task)
                self.completed_tasks.append(task)
                reward += 10 + (task["priority"] * 2)

        # Assign new task
        if self.tasks:
            task = self.tasks.pop(0)
            gpu_id = action
            self.task_waiting_times = [t + 1 for t in self.task_waiting_times]
            if (self.gpu_usage[gpu_id] + task["gpu_demand"] <= self.gpu_capacity[gpu_id] and
                self.memory_usage[gpu_id] + task["memory_demand"] <= self.memory_capacity[gpu_id]):
                task["gpu_id"] = gpu_id
                task["remaining_duration"] = task["duration"]
                task["start_time"] = self.total_time
                self.running_tasks.append(task)
                self.gpu_usage[gpu_id] += task["gpu_demand"]
                self.memory_usage[gpu_id] += task["memory_demand"]
                reward += (20 - self.task_waiting_times[0]) * 0.1
                self.task_waiting_times.pop(0)
            else:
                reward -= 2

        self.total_time += 1
        obs = np.array(self.gpu_usage + self.memory_usage, dtype=np.float32)
        done = len(self.tasks) == 0 and len(self.running_tasks) == 0

        if done:
            reward += -self.total_time * 0.1

        # Track GPU usage at this time step
        for i in range(self.num_gpus):
            self.gpu_usage_over_time[i].append(self.gpu_usage[i])

        return obs, reward, done, {}

    def render(self):
        print(f"GPU Usage: {self.gpu_usage}")
        print(f"Memory Usage: {self.memory_usage}")
        print(f"Remaining Tasks: {len(self.tasks)}")
        print(f"Total Time (Makespan): {self.total_time}")
        print(f"Average Waiting Time: {np.mean(self.task_waiting_times) if self.task_waiting_times else 0}")
        self._plot_gantt_chart()

    def _plot_gantt_chart(self):
        all_tasks = self.completed_tasks + self.running_tasks
        gpu_colors = list(mcolors.TABLEAU_COLORS.values())
        fig, ax = plt.subplots(figsize=(10, 6))
        for task in all_tasks:
            ax.barh(task["gpu_id"], task["duration"], left=task["start_time"],
                    color=gpu_colors[task["gpu_id"] % len(gpu_colors)], edgecolor="black", height=0.4)
            ax.text(task["start_time"] + task["duration"] / 2, task["gpu_id"], f'Task {task["id"]}',
                    va='center', ha='center', color='white', fontsize=8)

        ax.set_xlabel('Time')
        ax.set_ylabel('GPU')
        ax.set_title('Task Scheduling (Gantt Chart)')
        ax.set_yticks(range(self.num_gpus))
        ax.set_yticklabels([f'GPU {i}' for i in range(self.num_gpus)])
        ax.grid(True, axis='x', linestyle='--', alpha=0.7)
        plt.tight_layout()
        plt.show()


def evaluate_model(model, env, num_tasks=200):
    env.tasks = [generate_task(i) for i in range(num_tasks)]
    obs = env.reset()
    rewards = []
    total_waiting_time = 0

    while True:
        action, _ = model.predict(obs, deterministic=True)
        obs, reward, done, _ = env.step(action)
        rewards.append(reward)
        total_waiting_time += sum(env.task_waiting_times)
        env.render()
        if done:
            break

    avg_waiting_time = total_waiting_time / num_tasks
    print(f"Final Makespan (Total Time): {env.total_time}")
    print(f"Average Waiting Time: {avg_waiting_time:.2f}")
    print(f"Total Reward: {sum(rewards)}")
    return env.total_time, avg_waiting_time


def plot_gpu_usage(env):
    plt.figure(figsize=(10, 6))
    for i, usage in enumerate(env.gpu_usage_over_time):
        plt.plot(usage, label=f'GPU {i}')
    plt.xlabel('Time')
    plt.ylabel('GPU Usage (%)')
    plt.title('GPU Usage Over Time')
    plt.legend()
    plt.grid(True)
    plt.show()


# Create environment, train the model, and evaluate
env = MLWorkloadSchedulingEnv()
model = PPO("MlpPolicy", env, verbose=1)
model.learn(total_timesteps=50000)

# Evaluate and plot GPU usage
final_time, avg_waiting_time = evaluate_model(model, env)
plot_gpu_usage(env)
```
